```python
# create a fuction to find the factorial of a number
```

```python
def fact(n):
    factorial = 1

    for i in range(1, n + 1):
        factorial *= i

    print(factorial)
```

```python
fact(5)
```

```python
fact(10)
```

```python
# pass any number of integer in a fuction and find their sum
```

```python
def addall(*args):
    s = 0
    for i in args:
        s = s + i
        return s
```

```python
addall(10, 20, 30, 40, 5, 1, 5)
```
