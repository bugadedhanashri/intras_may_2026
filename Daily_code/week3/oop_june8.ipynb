```python
class Person:
    
    def __init__(self, name, age, ht):
        self.name = name
        self.age = age
        self.height = ht

    def show_info(self):
        return f"name -> {self.name}, age -> {self.age}, height -> {self.height}"
```

```python
p1 = Person('jatin',20,5.8)
p2 = Person('sachine', 22, 5.9)
```

```python
p1.show_info()
```

```python
"""
create a class called employee =>
attributes ->company,name ,monthly_salary,job_designation
here all the objects of this company will have the same common company
make salary as private variable 

create 3 objects -> e1,e2,e3 with given attributes 

method -> 1. method to calculate the anual salary
2. method to show entire data
""";
```

```python
class Employee:
    company = "ABC Technologies"   # Common for all employees

    def __init__(self, name, monthly_salary, job_designation):
        self.name = name
        self.__monthly_salary = monthly_salary  # Private variable
        self.job_designation = job_designation

    def annual_salary(self):
        print("Annual Salary =", self.__monthly_salary * 12)

    def info(self):
        return (f"Name -> {self.name}, "
                f"Company -> {Employee.company}, "
                f"Monthly Salary -> {self.__monthly_salary}, "
                f"Job Designation -> {self.job_designation}")
```

```python
e1 = Employee("Rahul", 50000, "Developer")
e2 = Employee("Priya", 60000, "Tester")
e3 = Employee("Amit", 70000, "Manager")
```

```python
print(e1.info())
e1.annual_salary()

print(e2.info())
e2.annual_salary()

print(e3.info())
e3.annual_salary()
```
