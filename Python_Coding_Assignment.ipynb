```python
 # 1. Take two numbers as input from the user and print their sum, difference, product, and quotient — each on a separate line. Handle division by zero with an appropriate message
```

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print("Sum =", num1 + num2)
print("Difference =", num1 - num2)
print("Product =", num1 * num2)

if num2 != 0:
    print("Quotient =", num1 / num2)
else:
    print("Error: Division by zero is not allowed.")
```

```python
#2. Accept a temperature in Celsius from the user, convert it to Fahrenheit and Kelvin, and print all three values with labels
```

```python
celsius = float(input("Enter temperature in Celsius: "))

fahrenheit = (celsius * 9/5) + 32
kelvin = celsius + 273.15

print("Celsius =", celsius)
print("Fahrenheit =", fahrenheit)
print("Kelvin =", kelvin)
```

```python
# 5 . Take a year as input and print whether it is a leap year or not.
```

```python
year = int(input("Enter a year"))
if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print("century leap")
        else:
            print("century but not leap")
    else:
        print("leap")
else:
    print("not leap")
    
```

```python
# 6. Build a simple calculator — take two numbers and an operator (+, -, *, /) as input,perform the operation, and print the result. Print an error message for invalid operators ordivision by zero
```

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operator = input("Enter operator (+, -, *, /): ")

if operator == '+':
    print(num1 + num2)

elif operator == '-':
    print(num1 - num2)

elif operator == '*':
    print(num1 * num2)

elif operator == '/':
    if num2 != 0:
        print(num1 / num2)
    else:
        print("Error: Division by zero")

else:
    print("Invalid operator")
```

```python
# 8. Take a positive integer n as input and print all prime numbers from 2 to n
```

```python
n = int(input("Enter a positive integer: "))

for num in range(2, n + 1):
    is_prime = True

    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num)
```

```python
count = 0
total = 0

while True:
    num = float(input("Enter a number (0 to stop): "))

    if num == 0:
        break

    total += num
    count += 1

if count != 0:
    average = total / count
    print("Count =", count)
    print("Sum =", total)
    print("Average =", average)
else:
    print("No numbers were entered.")
```

```python
# 10.. Take n as input and print a triangle pattern of n rows made of asterisks, where row i has (2i - 1) stars centered appropriately
```

```python
n = int(input("Enter number of rows: "))

for i in range(1, n + 1):
    spaces = n - i
    stars = 2 * i - 1
    print(" " * spaces + "*" * stars)
```
