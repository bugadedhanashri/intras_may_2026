```python
df = pd.read_csv('mtcars.csv')
df;
```

```python
df.columns
```

```python
x = df[['disp', 'cyl', 'wt']]
```

```python
y = df['mpg']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x, y)

lr.score(x, y) # r2 score 
```

```python
# RFE 
```

```python
from sklearn.feature_selection import RFE
```

```python
rfe = RFE(LinearRegression(), n_features_to_select= 3)
```

```python
x = df.drop('mpg', axis = 1)
```

```python
rfe.fit(x, y)
```

```python
rfe.get_feature_names_out()
```

```python
rfe.get_support()
```

```python
x_new = x.loc[:, rfe.get_support()]
x_new.columns
```

```python
rfe.score(x, y)
```
