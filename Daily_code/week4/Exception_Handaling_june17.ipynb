```python
# we need to be specific about kind of error we want to catch
```

```python
try:
    f = open('a1.txt')
    print(var1)
except FileNotFoundError:
    print("file does not exist")
```

```python
try:
    f = open('a1.txt')
    print(var1)
except FileNotFoundError:
    print("file does not exist")
except:
    print("something went wrong")
```

```python
try:
    f = open('a1.txt')
    print(var1)
except FileNotFoundError:
    print("file does not exist")
except Exception as e:
    print(e)
```

```python
# else -> if try block doesnt raise any exception this will run
```

```python
try:
    f = open('a1.txt')
    print(var1)
except FileNotFoundError:
    print("file does not exist")
except Exception as e:
    print(e)
else:
    print(f.read())
    f.close()
```

```python
# finally
```

```python
try:
    f = open('a1.txt')
    print(var1)
except FileNotFoundError:
    print("file does not exist")
except Exception as e:
    print(e)
else:
    print(f.read())
    f.close()
finally:
    print("run complete")
```

```python
# manually raise a exception
```

```python
try:
    f = open('a1.txt')
    if f.name == 'a1.txt':
        raise Exception
except FileNotFoundError:
    print("file does not exist")
except Exception as e:
    print(e)
else:
    print(f.read())
    f.close()
finally:
    print("run complete")
```

```python
try:
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))

    result = num1 / num2
    print("Division result =", result)

except ValueError:
    print("Error: Please enter valid numeric values.")

except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
```
