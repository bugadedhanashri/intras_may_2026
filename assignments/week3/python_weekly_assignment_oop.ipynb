```python
class BankAccount:
    def __init__(self, holder, number, balance):
        self.holder = holder
        self.number = number
        self.balance = balance
    def deposit(self, amount):
        self.balance += amount
    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient Balance")
        else:
            self.balance -= amount

    def get_balance(self):
        return self.balance

    def __str__(self):
        return f"{self.holder} , Nmuber: {self.number} , Balance: {self.balance}"
acc1 = BankAccount("Rahul", 101, 5000)
acc2 = BankAccount("Priya", 102, 8000)

acc1.deposit(1000)
acc1.withdraw(2000)

acc2.deposit(500)
acc2.withdraw(3000)

print(acc1)
print(acc2)
```

```python
### 2. Student Class
##Create a class Student with attributes:name,roll_no,marks (a list of 5 subject marks)
##Add methods:a) calculate_percentage() Returns the average of all marks. b) get_grade()
##'A' for percentage ≥ 80,'B' for percentage ≥ 60 ,'C' for percentage ≥ 40 'Fail' otherwise
##c) display() Prints all details including percentage and grade.
##Task: Create a list of 3 Student objects and call display() on each.
```

```python
class Student:
    def __init__(self, name, roll_no, marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks

    def calculate_percentage(self):
        return sum(self.marks) / len(self.marks)

    def get_grade(self):
        percentage = self.calculate_percentage()

        if percentage >= 80:
            return "A"
        elif percentage >= 60:
            return "B"
        elif percentage >= 35:
            return "C"
        else:
            return "Fail"

    def display(self):
        print("Name:", self.name)
        print("Roll No:", self.roll_no)
        print("Marks:", self.marks)
        print("Percentage:", self.calculate_percentage())
        print("Grade:", self.get_grade())
        print()


# Create 3 Student objects
s1 = Student("Dhanashri", 1, [85, 98, 88, 88, 92])
s2 = Student("Aarya", 2, [70, 65, 60, 75, 68])
s3 = Student("Amit", 3, [95, 90, 98, 92, 96])

students = [s1, s2, s3]

for student in students:
    student.display()
```

```python
##Q5. Shape Class
#Create a base class Shape with:A method area() that returns 0.A method describe() that prints
#"I am a shape".Create three subclasses: a) Circle(radius) — overrides area() using π × r².b) Rectangle
#(length, width) — overrides area().c) Triangle(base, height) — overrides area().
#Each subclass should also override describe() to print something specific about itself.Create one object
#of each class and call both methods.
```

```python
import math

class Shape:
    def area(self):
        return 0

    def describe(self):
        print("I am a shape")


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius * self.radius

    def describe(self):
        print("I am a Circle")


class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def describe(self):
        print("I am a Rectangle")


class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height

    def describe(self):
        print("I am a Triangle")


c = Circle(5)
r = Rectangle(4, 6)
t = Triangle(10, 8)

c.describe()
print("Area:", c.area())

r.describe()
print("Area:", r.area())

t.describe()
print("Area:", t.area())
```

```python
### Q6. Vehicle Class Question
#Create a class Vehicle with attributes make, model, and year and a method start_engine() that prints
#"Engine started".Create two subclasses:
#a) Car(make, model, year, num_doors) — overrides start_engine() to include the car's name in the message.
#b) ElectricCar(make, model, year, battery_capacity) — overrides start_engine() to print "Silent start —
#battery at X kWh".Use super() in both subclasses to call the parent constructor. Create objects and test 
#all methods.
```

```python
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def start_engine(self):
        print("Engine started")

class Car(Vehicle):
    def __init__(self, make, model, year, num_doors):
        super().__init__(make, model, year)
        self.num_doors = num_doors

    def start_engine(self):
        print(self.make, self.model, "Engine started")

class ElectricCar(Vehicle):
    def __init__(self, make, model, year, battery_capacity):
        super().__init__(make, model, year)
        self.battery_capacity = battery_capacity

    def start_engine(self):
        print("Silent start - battery at", self.battery_capacity, "kWh")

c1 = Car("Toyota", "Fortuner", 2024, 4)
e1 = ElectricCar("Tesla", "Model 3", 2024, 75)

c1.start_engine()
e1.start_engine()
```

```python
## Q7. Animal Class (Polymorphism)
#Question Create a base class Animal with a method sound() that prints "Some sound".
#Create subclasses Dog, Cat, and Cow, each overriding sound() with their respective sounds.
#Write a function make_all_sounds(animals) that takes a list of Animal objects and calls sound() on each.
#Demonstrate that the correct method is called regardless of the object type.
```

```python
class Animal:
    def sound(self):
        print("Some sound")


class Dog(Animal):
    def sound(self):
        print("Bark")


class Cat(Animal):
    def sound(self):
        print("Meow")


class Cow(Animal):
    def sound(self):
        print("Moo")


def make_all_sounds(animals):
    for animal in animals:
        animal.sound()


animals = [Dog(), Cat(), Cow()]
make_all_sounds(animals)
```

```python
## Q8. Calculator and Vector Class
#Question Create a class Calculator with overloaded-style methods using default arguments:
#a) add(a, b, c=0) — adds two or three numbers. b) multiply(a, b, c=1) — multiplies two or three numbers.
#c) Implement __add__ on a class Vector(x, y) so that two Vector objects can be added using the + operator.
#Print the resulting vector.
```

```python
class Calculator:
    def add(self, a, b, c=0):
        return a + b + c

    def multiply(self, a, b, c=1):
        return a * b * c


class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def display(self):
        print(self.x, self.y)


calc = Calculator()

print(calc.add(10, 20))
print(calc.add(10, 20, 30))

print(calc.multiply(2, 3))
print(calc.multiply(2, 3, 4))

v1 = Vector(2, 3)
v2 = Vector(4, 5)

v3 = v1 + v2
v3.display()
```
