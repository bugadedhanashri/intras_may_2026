````python
# Q1. Analyze a Sentence
# Question Write a function analyze_sentence(sentence) that takes a sentence and returns a dictionary with:word_count char_count (excluding spaces) unique_words (case-insensitive) most_frequent_word Test with:
````

````python
def analyze_sentence(sentence):
    words = sentence.lower().split()

    word_count = len(words)
    char_count = len(sentence.replace(" ", ""))
    unique_words = len(set(words))

    freq = {}
    for word in words:
        freq[word] = freq.get(word, 0) + 1

    most_frequent_word = max(freq, key=freq.get)

    return {
        "word_count": word_count,
        "char_count": char_count,
        "unique_words": unique_words,
        "most_frequent_word": most_frequent_word
    }
sentence = "The quick brown fox jumps over the lazy dog the fox"
print(analyze_sentence(sentence))
````

````python
#(b) Check if it is a palindrome (ignoring case and spaces)
````

````python
text = "A man a plan a canal Panama"

cleaned = text.replace(" ", "").lower()

is_palindrome = cleaned == "".join(reversed(cleaned))

print(is_palindrome)
````

````python
#(c) Replace every vowel with
````

````python
text = "A man a plan a canal Panama"

vowels = "aeiouAEIOU"
result = ""

for ch in text:
    if ch in vowels:
        result += "*"
    else:
        result += ch

print(result)
````

````python
text = "A man a plan a canal Panama"
vowel_count = {
    'a': 0,
    'e': 0,
    'i': 0,
    'o': 0,
    'u': 0
}
for ch in text.lower():
    if ch in vowel_count:
        vowel_count[ch] += 1
print(vowel_count)
````

````python
# Q4. Write a function list_ops(nums) that:

### a) Removes duplicates while preserving the original order.

### b) Returns a new list with only elements at even indices, squared.

### c) Flattens the nested list into a single list using list comprehension.
````

````python
def list_ops(nums):
    # a) Remove duplicates
    unique = []
    for num in nums:
        if num not in unique:
            unique.append(num)
    # b) Square elements at even indices
    squared_even = [nums[i] ** 2 for i in range(0, len(nums), 2)]
    return unique, squared_even
nums = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
nested = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]
flattened = [item for sublist in nested for item in sublist]
unique, squared = list_ops(nums)
print("Unique List:", unique)
print("Squared Even Index Elements:", squared)
print("Flattened List:", flattened)
````

````python
#Q7. Write a function inventory_manager(records) that takes a list of tuples (product, category, quantity, price) and returns a dictionary grouped by category. Each category should contain:*
### a) total_items
### b) total_value
### c) List of product names
### Given Data:
#python records = [("Laptop", "Electronics", 10, 55000),("Mouse", "Electronics", 50, 499),("Desk", "Furniture", 8, 12000),("Chair", "Furniture", 15, 4500),("Pen", "Stationery", 200, 20)]
````

````python
def inventory_manager(records):
    inventory = {}
    for product, category, quantity, price in records:
        if category not in inventory:
            inventory[category] = {
                "total_items": 0,
                "total_value": 0,
                "products": []
            }
        inventory[category]["total_items"] += quantity
        inventory[category]["total_value"] += quantity * price
        inventory[category]["products"].append(product)
    return inventory
# Given Data
records = [("Laptop", "Electronics", 10, 55000),("Mouse", "Electronics", 50, 499),
("Desk", "Furniture", 8, 12000),("Chair", "Furniture", 15, 4500),("Pen", "Stationery", 200, 20)]
result = inventory_manager(records)
for category, details in result.items():
    print(category, ":", details)
````

````python
#(a) Function: power_all(*args, exp=2)

##Raises each number to the given exponent (default = 2).
````

````python
def power_all(*args, exp=2):
    return [x ** exp for x in args]

# Example:
print(power_all(2, 3, 4))          # [4, 9, 16]
print(power_all(2, 3, 4, exp=3))   # [8, 27, 64]
````

````python
#1. Module: stats_utils.py
````

````python
import math
def mean(data):
    return sum(data) / len(data)
def variance(data):
    m = mean(data)
    return sum((x - m) ** 2 for x in data) / len(data)  # population variance
def normalize(data):
    min_val = min(data)
    max_val = max(data)
    if max_val == min_val:
        return [0 for _ in data]  # avoid division by zero
    return [(x - min_val) / (max_val - min_val) for x in data]
````

````python
### Q10. Complete the following:
### a) Write a closure make_multiplier(n) that returns a function multiplying its input by n. Create triple and quintuple using it.
### b) Write a decorator @timer that prints the execution time of any function it wraps. Apply it to slow_sum(n) which computes the sum of squares from 1 to n. 
### c) Write a decorator @validate_positive that raises a ValueError if any argument is zero or negative. Apply it to safe_sqrt(x).
````

````python
###a) Closure: make_multiplier(n)
````

````python
def make_multiplier(n):
    def multiplier(x):
        return x * n
    return multiplier

triple = make_multiplier(3)
quintuple = make_multiplier(5)

print("Triple of 10:", triple(10))
print("Quintuple of 10:", quintuple(10))
````

````python
##b) Decorator: @timer
````

````python
import time

def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()

        print(f"Execution Time: {end - start:.6f} seconds")
        return result
    return wrapper


@timer
def slow_sum(n):
    return sum(i ** 2 for i in range(1, n + 1))

print("Sum of Squares:", slow_sum(100000))
````

````python
##c) Decorator: @validate_positive
````

````python
import math

def validate_positive(func):
    def wrapper(*args, **kwargs):
        for arg in args:
            if arg <= 0:
                raise ValueError("Arguments must be positive.")
        return func(*args, **kwargs)
    return wrapper


@validate_positive
def safe_sqrt(x):
    return math.sqrt(x)

print("Square Root:", safe_sqrt(25))
````
