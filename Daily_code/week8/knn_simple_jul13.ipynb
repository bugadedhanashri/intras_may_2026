```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('ingredients.csv')
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
sns.jointplot( x='sweetness' , y ='crunchiness', data = df, hue = 'class');
```

```python
# seprate input and output
```

```python
x = df[['sweetness' , 'crunchiness']]
y = df['class']
```

```python
# build the model
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 4)

knn.fit(x,y)
```

```python
tomato = pd.DataFrame({'sweetness':[6],
                       'crunchiness':[4]})
```

```python
knn.predict(tomato)
```
