```python
class Person:
    insti = "MITU"

    def __init__ (self, age, ht):
        self.age = age
        self.ht = ht

    def show(self):
        print(f"age -> {self.age}, height -> {self.ht}, institute -> {self.insti}")
```

```python
class Ravi(Person):

    def __init__(self, age, ht, sal):
        super().__init__(age, ht)
        self.sal = sal

    def show_sal(self):
        print(f"salary is {self.sal}")
```

```python
a = Ravi(25, 5.9, 5700)
```

```python
a.show_sal()
```

```python
a.show()
```

```python
class Menuapp:

    def __init__ (self):
        self.itemlist = []

    def show_menu(self):

        print("Menu\n1.add items\n2. view items\n3. exit")

    def add_items(self):
        item = input("enter the item name:")
        self.itemlist.append(item)

    def view_items(self):
        return self.itemlist
    
    def run(self):

        while True:
            
            self.show_menu()

            ch = int(input("enter the choice"))

            if ch == 1:
                self.add_items()
            elif ch == 2:
                self.view_items()
            elif ch == 3:
                break
            else:
                return 'invalid choice'
```

```python
atharv = Menuapp()
```

```python
atharv.run()
```

```python
atharv.itemlist
```

```python
class Person:

    def __init__ (self, age, name):

        self.__age = age
        self.__name = name

    # getter
    def get_name(self):
        return self.__name
    
    def get_age(self):
        return self.__age
    
    # setter
    def set_age(self, age):
        self.__age = age
        

    def __str__(self):
        return f"age -> {self.__age}, name -> {self.__name}"
```

```python
# create a menu driven app for banking simulation
# check balance, deposit money, withdraw money (u can add more things as u see fit to make the app better)
```

```python
class BankAccount:
    def __init__(self, initial_balance=0, pin="1234"):
        self.balance = initial_balance
        self.pin = pin
        self.transactions = []

    def check_balance(self):
        print(f"\nCurrent Balance: ₹{self.balance:.2f}")

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            self.transactions.append(f"Deposited ₹{amount:.2f}")
            print(f"₹{amount:.2f} deposited successfully.")
        else:
            print("Invalid deposit amount.")

    def withdraw(self, amount):
        if amount <= 0:
            print("Invalid withdrawal amount.")
        elif amount > self.balance:
            print("Insufficient balance.")
        else:
            self.balance -= amount
            self.transactions.append(f"Withdrawn ₹{amount:.2f}")
            print(f"₹{amount:.2f} withdrawn successfully.")

    def transfer(self, amount, recipient):
        if amount <= 0:
            print("Invalid transfer amount.")
        elif amount > self.balance:
            print("Insufficient balance.")
        else:
            self.balance -= amount
            self.transactions.append(
                f"Transferred ₹{amount:.2f} to {recipient}"
            )
            print(f"₹{amount:.2f} transferred to {recipient} successfully.")

    def show_transactions(self):
        print("\n--- Transaction History ---")
        if not self.transactions:
            print("No transactions found.")
        else:
            for i, transaction in enumerate(self.transactions, start=1):
                print(f"{i}. {transaction}")

    def change_pin(self):
        old_pin = input("Enter current PIN: ")
        if old_pin == self.pin:
            new_pin = input("Enter new PIN: ")
            self.pin = new_pin
            print("PIN changed successfully.")
        else:
            print("Incorrect PIN.")


def login(account):
    attempts = 3

    while attempts > 0:
        entered_pin = input("Enter PIN: ")
        if entered_pin == account.pin:
            print("\nLogin successful!")
            return True
        else:
            attempts -= 1
            print(f"Incorrect PIN. Attempts left: {attempts}")

    print("Account locked. Too many failed attempts.")
    return False


def main():
    account = BankAccount(initial_balance=5000)

    if not login(account):
        return

    while True:
        print("\n===== BANKING SYSTEM =====")
        print("1. Check Balance")
        print("2. Deposit Money")
        print("3. Withdraw Money")
        print("4. Transfer Money")
        print("5. Transaction History")
        print("6. Change PIN")
        print("7. Exit")

        choice = input("Enter your choice (1-7): ")

        if choice == "1":
            account.check_balance()

        elif choice == "2":
            amount = float(input("Enter amount to deposit: ₹"))
            account.deposit(amount)

        elif choice == "3":
            amount = float(input("Enter amount to withdraw: ₹"))
            account.withdraw(amount)

        elif choice == "4":
            recipient = input("Enter recipient name: ")
            amount = float(input("Enter amount to transfer: ₹"))
            account.transfer(amount, recipient)

        elif choice == "5":
            account.show_transactions()

        elif choice == "6":
            account.change_pin()

        elif choice == "7":
            print("Thank you for banking with us!")
            break

        else:
            print("Invalid choice. Please try again.")


if __name__ == "__main__":
    main()
```
