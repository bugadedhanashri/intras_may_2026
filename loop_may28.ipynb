```python
# write program to find factorial
```

```python
num = int(input("enter the number"))
fact = 1

while num > 0:
    fact = fact * num
    num -= 1

print(fact)
```

```python
# hint -> make use of % and //
```

```python
num = 12345

add = 0

while num > 0:
    rem = num % 10   
    add = add + rem    
    num = num // 10    

print(add)
```

```python
num = int(input("enter the number"))
fact = 1

for n in range(1, num + 1):
    fact = fact * n

print(fact)
```
