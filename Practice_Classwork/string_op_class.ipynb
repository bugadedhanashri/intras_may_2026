```python
# 1 Write a Python function that takes a string as input. Using a for loop, iterate through each character. If the character is a vowel (a, e, i, o, u, case-insensitive), increment a counter. Store the count of each vowel in a dictionary. Finally, print the total number of vowels found and the dictionary of vowel counts.
```

```python
def count_vowels(text):
    vowel_counts = {'a': 0, 'e': 0, 'i': 0, 'o': 0, 'u': 0}
    total_vowels = 0

    for char in text:
        lower_char = char.lower()
        if lower_char in vowel_counts:
            vowel_counts[lower_char] += 1
            total_vowels += 1

    print("Total number of vowels:", total_vowels)
    print("Vowel counts:", vowel_counts)
```

```python
count_vowels("Dhanashri Bugade")
```

```python
# 2 .Palindrome Checker with List: Write a Python function that accepts a string. Inside the function, use a for loop to build a new list containing the characters of the string in reverse order. Then, use an if-else statement to check if the original string is a palindrome (reads the same forwards and backwards) by comparing it to the string formed from the reversed list. Print whether the string is a palindrome or not.
```

```python
def palindrome_checker(text):
    reversed_list = []

    for i in range(len(text) - 1, -1, -1):
        reversed_list.append(text[i])

    reversed_text = "".join(reversed_list)

    if text == reversed_text:
        print(f'"{text}" is a palindrome.')
    else:
        print(f'"{text}" is not a palindrome.')
```

```python
palindrome_checker("Amit")
```

```python
palindrome_checker("madam")
```

```python
# 3.Character Frequency Counter:Create a Python function that takes a string. Use a for loop to iterate through the string and populate a dictionary where keys are characters and values are their frequencies. Use an if-else statement to handle whether a character is already in the dictionary or not. Print the resulting character frequency dictionary.
```

```python
def character_frequency(text):
    freq = {}

    for char in text:
        if char in freq:
            freq[char] += 1
        else:
            freq[char] = 1

    print("Character Frequency Dictionary:")
    print(freq)
```

```python
character_frequency("hello")
```

```python
# 4 Word Length Classification:Write a Python function that takes a sentence (string) as input. Split the sentence into words. Using a for loop, iterate through each word. Use if-elif-else statements to classify words as "short" (length < 5), "medium" (5 <= length <= 8), or "long" (length > 8). Store these classified words in separate lists (e.g., short_words, medium_words, long_words). Print all three lists.
```

```python
def classify_words(sentence):
    words = sentence.split()

    short_words = []
    medium_words = []
    long_words = []

    for word in words:
        if len(word) < 5:
            short_words.append(word)
        elif 5 <= len(word) <= 8:
            medium_words.append(word)
        else:
            long_words.append(word)

    print("Short words:", short_words)
    print("Medium words:", medium_words)
    print("Long words:", long_words)
```

```python
classify_words("Python is a powerful programming language for beginners")
```

```python
# 5 Substrings with Specific Start/End: Given a list of strings, write a function that iterates through each string. If a string starts with a specific prefix (e.g., "py") and ends with a specific suffix (e.g., "thon"), add it to a new list called matching_strings. Use an if statement to check the conditions. Print the matching_strings list.
```

```python
def find_matching_strings(strings):
    matching_strings = []
    for word in strings:
        if word.startswith("py") and word.endswith("thon"):
            matching_strings.append(word)
    print("Matching Strings:", matching_strings)
```

```python
words = ["python", "pythons", "pyathon", "py123thon", "java", "pycoolthon"]
find_matching_strings(words)
```

```python
# 6 Unique Word Extractor: Write a function that takes a paragraph (string) as input. Convert the paragraph to lowercase and remove punctuation. Use a for loop to split the paragraph into words and add each unique word to a set data structure. Finally, convert the set back to a sorted list and print it.
```

```python
import string

def unique_words(paragraph):
    paragraph = paragraph.lower()
    for punct in string.punctuation:
        paragraph = paragraph.replace(punct, "")
    unique_set = set()
    for word in paragraph.split():
        unique_set.add(word)
    sorted_words = sorted(list(unique_set))
    print("Unique Words (Sorted):")
    print(sorted_words)
```

```python
text = "Hello, world! Hello Python. Python is great, and the world is big."
unique_words(text)
```

```python
# 7 .Even/Odd Indexed Character Separator: Create a Python function that takes a string. Using a for loop with enumerate, iterate through the string. Use an if-else statement to check if the index of a character is even or odd. Store characters at even indices in one list and characters at odd indices in another list. Print both lists.
```

```python
def even_odd_index_separator(text):
    even_chars = []
    odd_chars = []
    for index, char in enumerate(text):
        if index % 2 == 0:
            even_chars.append(char)
        else:
            odd_chars.append(char)
    print("Even index characters:", even_chars)
    print("Odd index characters:", odd_chars)
```

```python
even_odd_index_separator("PythonProgramming")
```

```python
# 8 .String Anagram Checker: Write a function that takes two strings as input. Convert both strings into lists of characters. Sort both lists. Use an if-else statement to check if the sorted lists are identical. Print whether the two strings are anagrams of each other (contain the same characters with the same frequencies).
```

```python
def anagram_checker(str1, str2):
    list1 = list(str1)
    list2 = list(str2)

    list1.sort()
    list2.sort()

    if list1 == list2:
        print("The strings are anagrams of each other.")
    else:
        print("The strings are NOT anagrams of each other.")
```

```python
anagram_checker("listen", "silent")
anagram_checker("hello", "world")
```

```python
# 9 .Missing Character Finder (Alphabet Check): Given a string, write a function that determines which lowercase English alphabet letters are not present in the string. Use a set to store all lowercase letters. Iterate through the input string, removing each found letter from the set. Print the remaining letters in the set.
```

```python
def missing_characters(text):
    alphabet = set("abcdefghijklmnopqrstuvwxyz")

    text = text.lower()

    for char in text:
        if char in alphabet:
            alphabet.remove(char)

    print("Missing characters:", alphabet)
```

```python
missing_characters("The quick brown fox jumps")
```
