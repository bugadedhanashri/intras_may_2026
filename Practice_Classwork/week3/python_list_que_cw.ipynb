```python
##Question 2: Sum of Even Numbers

#Write Python code that iterates through the following list numbers = [12, 7, 23, 48, 15, 66, 9].
#Calculate and print the sum of all even numbers in the list.
```

```python
numbers = [12, 7, 23, 48, 15, 66, 9]
sum_even = 0
for i in numbers:
    if i % 2 == 1:
        print(i, end = ' ')
        sum_even = sum_even + i
        print("sum of even number" ,sum_even)
```

```python
## Question 3: Manual List Reversal

#Given a list data = ['apple', 'banana', 'cherry', 'date', 'elderberry'], create a new list that contains
#the elements of data in reverse order. You are not allowed to use data.reverse() or data[::-1]. You must
#achieve this using loops and basic list operations.
```

```python
## Question 4: Finding the Maximum Element

#Without using the built-in max() 
#function, write code to find and print the largest number in the list values = [8, 2, 10, 5, 17, 3, 14].
```

```python
values = [8, 2, 10, 5, 17, 3, 14]

largest = 0

for i in values:
    if i > largest:
        largest = i

print(largest)
```

```python
## Question 5: Count Occurrences

#Given the list items = ['red', 'blue', 'green', 'red', 'yellow', 'blue', 'red'], write code to count how
#many times the string 'red' appears in the list and print the count. Do not use the list.count() method.
```

```python
items = ['red', 'blue', 'green', 'red', 'yellow', 'blue', 'red']
count = 0
for i in items:
    if i=="red":
        count = count + 1
print("count",count)
```

```python
## Question 6: Remove All Occurrences of an Element

#Given a list mixed_numbers = [1, 5, 2, 5, 3, 5, 4, 5], write code to remove all instances 
# of the number 5 from the list. The final list should only contain [1, 2, 3, 4].
```

```python
mixed_numbers = [1, 5, 2, 5, 3, 5, 4, 5]

while 5 in mixed_numbers:
    mixed_numbers.remove(5)

print(mixed_numbers)
```

```python
##m Question 9: Flatten a Nested List
#Given a nested list: nested_list = [1, [2, 3], 4, [5, 6, 7], 8]
#Write a program to convert this nested list into a single flat list (only one level of nesting is
#present).
```

```python
nested_list = [1, [2, 3], 4, [5, 6, 7], 8]

flat_list = []

for item in nested_list:
    if type(item) == list:
        for num in item:
            flat_list.append(num)
    else:
        flat_list.append(item)

print(flat_list)
```
